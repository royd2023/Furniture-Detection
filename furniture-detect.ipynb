```python
from ultralytics import YOLO
model = YOLO("yolo11n.pt")  # Load a pretrained model
results = model.predict(source='https://ultralytics.com/images/zidane.jpg', save=True)  # Predict on an image or video
print(results)  # Print results to console
```

```python
from PIL import Image
from IPython.display import display

results = model('https://ultralytics.com/images/zidane.jpg')  # Predict

for r in results:
    im_array = r.plot()  # Draw boxes on image as numpy array
    im = Image.fromarray(im_array)
    display(im)
```
